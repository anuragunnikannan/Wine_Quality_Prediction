## Imports

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## Loading Dataset

```python
df = pd.read_csv("winequality-red.csv")
```

```python
df.head()
```

```python
df.info()
# so there is no null values
```

```python
df.describe()
```

## Exploratory Data Analysis

```python
# checking number of different types of quality of wines
df['quality'].value_counts()

# poor and best quality wines are less and mediocre quality wines are more
```

```python
sns.countplot(df['quality'])
```

```python
# quality 3 and 4 are labelled as 0, 5 and 6 as 1, 7 and 8 as 2

label = []
for i in df['quality']:
    if i == 3 or i == 4:
        label.append(0)
    elif i == 5 or i == 6:
        label.append(1)
    elif i == 7 or i == 8:
        label.append(2)

df['label'] = label
```

```python
df.head()
```

```python
correlation = df.corr()['label'].sort_values(ascending=False)
correlation
```

```python
# Checking correlation between all the features
plt.figure(figsize=(10, 6))
sns.heatmap(df.corr(), cmap='Blues', annot=True, fmt='.2f')
```

```python
# We are taking features that have correlation value of more than 0.2 between quality and other features
print(abs(correlation) > 0.2)
```

### Sulphate content across different qualities of wine

## Feature Selection

```python
x = df[['alcohol', 'sulphates', 'volatile acidity', 'citric acid']]
y = df['label']
```

## Data Preprocessing

```python
from imblearn.over_sampling import SMOTE
```

```python
oversampling = SMOTE()
x_res, y_res = oversampling.fit_resample(x, y)
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x_res, y_res, test_size=0.2, random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
sc = StandardScaler()
```

```python
x_train = sc.fit_transform(x_train)
x_test = sc.transform(x_test)
```

```python
model_accuracy = {
    'Logistic Regression': [],
    'Decision Tree Classifier': [],
    'Random Forest Classifier': [],
    'Support Vector Classifier': [],
    'KNN Classifier': [],
    'Extra Trees Classifier': []
}
```

## Logisitic Regression

```python
from sklearn.linear_model import LogisticRegression
```

```python
log_reg = LogisticRegression()
```

```python
log_reg.fit(x_train, y_train)
```

```python
y_pred = log_reg.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score, recall_score, precision_score, f1_score
```

```python
model_accuracy['Logistic Regression'].append(accuracy_score(y_test, y_pred))
print(accuracy_score(y_test, y_pred))
```

```python
print(confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred))
```

## Decision Tree Classifier

```python
dtc = DecisionTreeClassifier()
```

```python
dtc.fit(x_train, y_train)
```

```python
y_pred = dtc.predict(x_test)
```

```python
model_accuracy['Decision Tree Classifier'].append(accuracy_score(y_test, y_pred))
print(accuracy_score(y_test, y_pred))
```

```python
print(confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred))
```
